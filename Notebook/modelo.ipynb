
El cÃ¡lculo de similitud del coseno es una medida comÃºnmente utilizada en el filtrado colaborativo para comparar la similitud entre vectores de caracterÃ­sticas, como las preferencias de los usuarios en un sistema de recomendaciÃ³n. En el contexto del modelo de recomendaciones basado en filtrado colaborativo basado en usuarios, el coseno de la similitud se utiliza para medir cuÃ¡n parecidas son las preferencias de dos usuarios en un espacio de alta dimensiÃ³n, donde cada dimensiÃ³n representa un Ã­tem (en este caso, un negocio).

El cÃ¡lculo de la similitud del coseno ayuda al modelo de recomendaciones a identificar usuarios con gustos similares. Esto es fundamental para el modelo, ya que se basa en la premisa de que si dos usuarios tienen gustos similares para ciertos Ã­tems, es probable que tambiÃ©n tengan gustos similares para otros Ã­tems. Al identificar usuarios similares, el modelo puede recomendar Ã­tems que hayan sido apreciados por estos usuarios similares pero que el usuario en cuestiÃ³n aÃºn no haya experimentado.

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
import pandas as pd

def get_recommendations(user_id, categoria):
    # Importa dataset
    df =  pd.read_csv('../docs/data_modelo/df_recomendaciones.csv.gz')
    # Obtener la ciudad y el estado del usuario
    ciudad_usuario = df[df['user_id'] == user_id]['city'].iloc[0]
    estado_usuario = df[df['user_id'] == user_id]['state'].iloc[0]

    # Filtrar el DataFrame para obtener solo las filas con la categorÃ­a especificada
    df_categoria = df[(df[categoria] == 1) & ((df['city'] == ciudad_usuario) | (df['state'] == estado_usuario))]

    if len(df_categoria) == 0:
        # Si no hay negocios en la categorÃ­a y ciudad o estado especificados, obtener las 3 mejores calificaciones y mÃ¡s similares
        matriz_calificaciones = df.pivot_table(index='user_id', columns='name', values='stars', fill_value=0)
        scaler = StandardScaler()
        matriz_calificaciones_escaladas = scaler.fit_transform(matriz_calificaciones)
        similitud_usuarios = cosine_similarity(matriz_calificaciones_escaladas)
        indice_usuario = matriz_calificaciones.index.get_loc(user_id)
        usuarios_similares = np.argsort(similitud_usuarios[indice_usuario])[::-1][1:]
        calificaciones_similares = matriz_calificaciones.iloc[usuarios_similares]
        calificaciones_promedio = calificaciones_similares.mean(axis=0)
        negocios_recomendados = calificaciones_promedio.sort_values(ascending=False).index.tolist()
        return negocios_recomendados[:3]

    # Convertir el DataFrame a una matriz de calificaciones de usuarios y negocios
    matriz_calificaciones = df_categoria.pivot_table(index='user_id', columns='name', values='stars', fill_value=0)

    # Escalar las calificaciones para que tengan media 0 y desviaciÃ³n estÃ¡ndar 1
    scaler = StandardScaler()
    matriz_calificaciones_escaladas = scaler.fit_transform(matriz_calificaciones)

    # Calcular la similitud de coseno entre usuarios
    similitud_usuarios = cosine_similarity(matriz_calificaciones_escaladas)

    # Obtener el Ã­ndice del usuario dado en la matriz de similitud
    indice_usuario = matriz_calificaciones.index.get_loc(user_id)

    # Obtener los usuarios mÃ¡s similares al usuario dado
    usuarios_similares = np.argsort(similitud_usuarios[indice_usuario])[::-1][1:]  # Excluimos al propio usuario

    # Obtener las calificaciones de los usuarios similares
    calificaciones_similares = matriz_calificaciones.iloc[usuarios_similares]

    # Calcular la calificaciÃ³n promedio para cada negocio
    calificaciones_promedio = calificaciones_similares.mean(axis=0)

    # Obtener los nombres de los negocios recomendados, ordenados por calificaciÃ³n promedio
    negocios_recomendados = calificaciones_promedio.sort_values(ascending=False).index.tolist()

    return negocios_recomendados[:3]  # Devolver las 3 mejores recomendaciones
```

```python
# Ejemplo de uso
user_id = '4Uh27DgGzsp6PqrH913giQ'
categoria = 'Restaurants'
sugerencias = get_recommendations(user_id, categoria)

if sugerencias:
    print("Sugerencias basadas en la categorÃ­a '{}' para el usuario '{}':".format(categoria, user_id))
    for i, sugerencia in enumerate(sugerencias):
        print(f"{i+1}. {sugerencia}")
else:
    print("No hay negocios recomendados para la categorÃ­a '{}' para el usuario '{}'.".format(categoria, user_id))
```

El modelo de recomendaciÃ³n basado en filtrado colaborativo basado en usuarios se selecciona debido a su capacidad para ofrecer recomendaciones personalizadas y relevantes utilizando Ãºnicamente informaciÃ³n sobre las calificaciones de los usuarios. Este enfoque es adecuado cuando se dispone de un conjunto de datos que contiene interacciones explÃ­citas entre usuarios y elementos (en este caso, calificaciones de negocios por parte de los usuarios) y se busca recomendar elementos a usuarios en funciÃ³n de la similitud con otros usuarios.

El modelo utiliza la similitud de los usuarios para identificar a aquellos con gustos similares y recomendar elementos que hayan sido apreciados por estos usuarios similares pero que el usuario en cuestiÃ³n aÃºn no haya experimentado. Esto lo hace especialmente Ãºtil en escenarios donde se desea ofrecer recomendaciones personalizadas sin necesidad de informaciÃ³n adicional sobre los usuarios o los elementos.

AdemÃ¡s, el modelo es relativamente fÃ¡cil de implementar y entender, lo que lo hace adecuado para aplicaciones prÃ¡cticas. Sin embargo, es importante tener en cuenta que el filtrado colaborativo basado en usuarios puede verse afectado por problemas como el problema de arranque en frÃ­o y la escalabilidad en conjuntos de datos muy grandes, por lo que se deben considerar otras tÃ©cnicas como el filtrado colaborativo basado en Ã­tems o en contenido en esos casos.

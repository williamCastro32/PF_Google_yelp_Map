
El cálculo de similitud del coseno es una medida comúnmente utilizada en el filtrado colaborativo para comparar la similitud entre vectores de características, como las preferencias de los usuarios en un sistema de recomendación. En el contexto del modelo de recomendaciones basado en filtrado colaborativo basado en usuarios, el coseno de la similitud se utiliza para medir cuán parecidas son las preferencias de dos usuarios en un espacio de alta dimensión, donde cada dimensión representa un ítem (en este caso, un negocio).

El cálculo de la similitud del coseno ayuda al modelo de recomendaciones a identificar usuarios con gustos similares. Esto es fundamental para el modelo, ya que se basa en la premisa de que si dos usuarios tienen gustos similares para ciertos ítems, es probable que también tengan gustos similares para otros ítems. Al identificar usuarios similares, el modelo puede recomendar ítems que hayan sido apreciados por estos usuarios similares pero que el usuario en cuestión aún no haya experimentado.

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
import pandas as pd

def get_recommendations(user_id, categoria):
    # Importa dataset
    df =  pd.read_csv('../docs/data_modelo/df_recomendaciones.csv.gz')
    # Obtener la ciudad y el estado del usuario
    ciudad_usuario = df[df['user_id'] == user_id]['city'].iloc[0]
    estado_usuario = df[df['user_id'] == user_id]['state'].iloc[0]

    # Filtrar el DataFrame para obtener solo las filas con la categoría especificada
    df_categoria = df[(df[categoria] == 1) & ((df['city'] == ciudad_usuario) | (df['state'] == estado_usuario))]

    if len(df_categoria) == 0:
        # Si no hay negocios en la categoría y ciudad o estado especificados, obtener las 3 mejores calificaciones y más similares
        matriz_calificaciones = df.pivot_table(index='user_id', columns='name', values='stars', fill_value=0)
        scaler = StandardScaler()
        matriz_calificaciones_escaladas = scaler.fit_transform(matriz_calificaciones)
        similitud_usuarios = cosine_similarity(matriz_calificaciones_escaladas)
        indice_usuario = matriz_calificaciones.index.get_loc(user_id)
        usuarios_similares = np.argsort(similitud_usuarios[indice_usuario])[::-1][1:]
        calificaciones_similares = matriz_calificaciones.iloc[usuarios_similares]
        calificaciones_promedio = calificaciones_similares.mean(axis=0)
        negocios_recomendados = calificaciones_promedio.sort_values(ascending=False).index.tolist()
        return negocios_recomendados[:3]

    # Convertir el DataFrame a una matriz de calificaciones de usuarios y negocios
    matriz_calificaciones = df_categoria.pivot_table(index='user_id', columns='name', values='stars', fill_value=0)

    # Escalar las calificaciones para que tengan media 0 y desviación estándar 1
    scaler = StandardScaler()
    matriz_calificaciones_escaladas = scaler.fit_transform(matriz_calificaciones)

    # Calcular la similitud de coseno entre usuarios
    similitud_usuarios = cosine_similarity(matriz_calificaciones_escaladas)

    # Obtener el índice del usuario dado en la matriz de similitud
    indice_usuario = matriz_calificaciones.index.get_loc(user_id)

    # Obtener los usuarios más similares al usuario dado
    usuarios_similares = np.argsort(similitud_usuarios[indice_usuario])[::-1][1:]  # Excluimos al propio usuario

    # Obtener las calificaciones de los usuarios similares
    calificaciones_similares = matriz_calificaciones.iloc[usuarios_similares]

    # Calcular la calificación promedio para cada negocio
    calificaciones_promedio = calificaciones_similares.mean(axis=0)

    # Obtener los nombres de los negocios recomendados, ordenados por calificación promedio
    negocios_recomendados = calificaciones_promedio.sort_values(ascending=False).index.tolist()

    return negocios_recomendados[:3]  # Devolver las 3 mejores recomendaciones
```

```python
# Ejemplo de uso
user_id = '4Uh27DgGzsp6PqrH913giQ'
categoria = 'Restaurants'
sugerencias = get_recommendations(user_id, categoria)

if sugerencias:
    print("Sugerencias basadas en la categoría '{}' para el usuario '{}':".format(categoria, user_id))
    for i, sugerencia in enumerate(sugerencias):
        print(f"{i+1}. {sugerencia}")
else:
    print("No hay negocios recomendados para la categoría '{}' para el usuario '{}'.".format(categoria, user_id))
```

El modelo de recomendación basado en filtrado colaborativo basado en usuarios se selecciona debido a su capacidad para ofrecer recomendaciones personalizadas y relevantes utilizando únicamente información sobre las calificaciones de los usuarios. Este enfoque es adecuado cuando se dispone de un conjunto de datos que contiene interacciones explícitas entre usuarios y elementos (en este caso, calificaciones de negocios por parte de los usuarios) y se busca recomendar elementos a usuarios en función de la similitud con otros usuarios.

El modelo utiliza la similitud de los usuarios para identificar a aquellos con gustos similares y recomendar elementos que hayan sido apreciados por estos usuarios similares pero que el usuario en cuestión aún no haya experimentado. Esto lo hace especialmente útil en escenarios donde se desea ofrecer recomendaciones personalizadas sin necesidad de información adicional sobre los usuarios o los elementos.

Además, el modelo es relativamente fácil de implementar y entender, lo que lo hace adecuado para aplicaciones prácticas. Sin embargo, es importante tener en cuenta que el filtrado colaborativo basado en usuarios puede verse afectado por problemas como el problema de arranque en frío y la escalabilidad en conjuntos de datos muy grandes, por lo que se deben considerar otras técnicas como el filtrado colaborativo basado en ítems o en contenido en esos casos.
